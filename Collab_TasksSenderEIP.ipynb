```python

# Referencia a la colección 'tasks'
tasks_ref = db.collection('tasks')

# Buscar todas las tareas con _id >= 175
tasks_to_delete = tasks_ref.where('_id', '>=', 175).stream()

# Eliminar cada tarea encontrada
deleted_count = 0
for task in tasks_to_delete:
    task.reference.delete()
    deleted_count += 1
    print(f"Tarea con ID {task.id} eliminada")

print(f"Se eliminaron {deleted_count} tareas con _id >= 175.")
```
