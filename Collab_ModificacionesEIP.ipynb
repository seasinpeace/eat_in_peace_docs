```python
def find_duplicate_nameEn_in_food():
    try:
        # Referencia a la colección 'food'
        food_ref = db.collection('food')
        docs = food_ref.stream()

        # Diccionario para contar las repeticiones de nameEn
        name_counts = {}

        for doc in docs:
            data = doc.to_dict()
            nameEn = data.get('nameEn')

            if nameEn:
                if nameEn in name_counts:
                    name_counts[nameEn] += 1
                else:
                    name_counts[nameEn] = 1

        # Filtrar los nameEn que aparecen más de una vez
        duplicates = {name: count for name, count in name_counts.items() if count > 1}

        # Imprimir y devolver los resultados
        for name, count in duplicates.items():
            print(f"nameEn: {name}, count: {count}")

        return duplicates

    except Exception as e:
        print(f"Error al buscar duplicados: {e}")
        return {}

# Llamar a la función para encontrar duplicados
duplicate_nameEn = find_duplicate_nameEn_in_food()
```

```python
def calculate_kcal_for_recipes():
    try:
        # Referencia a las colecciones
        recipes_ref = db.collection('recipes')
        recipes_ing_ref = db.collection('recipes_ing')
        food_ref = db.collection('food')
        utils_cons_ref = db.collection('utils_cons')

        # Obtener todas las recetas
        recipes_docs = recipes_ref.stream()

        for recipe_doc in recipes_docs:
            recipe_data = recipe_doc.to_dict()
            recipe_id = recipe_data.get('_id')

            # Si la receta ya tiene el campo kcal, continuar con la siguiente
            if recipe_data.get('kcal') is not None:
                continue

            print(f"Calculando kcal para receta con _id = {recipe_id}")

            # Obtener los ingredientes de esta receta
            ingredients_query = recipes_ing_ref.where('id_recipe', '==', recipe_id).stream()
            kcal_total = 0

            for ing_doc in ingredients_query:
                ing_data = ing_doc.to_dict()
                amount = ing_data.get('amount', 0)
                unit = ing_data.get('unit')
                item_id = ing_data.get('item_id')

                # Validar que el campo amount sea convertible a float
                try:
                    amount = float(amount) if amount else 0
                except ValueError:
                    print(f"Valor no válido para 'amount': {amount}")
                    continue

                if unit == 'g':
                    # Buscar en food para obtener kcal_100g
                    food_query = food_ref.where('_id', '==', item_id).stream()
                    for food_doc in food_query:
                        food_data = food_doc.to_dict()
                        kcal_100g = food_data.get('kcal_100g', 0)
                        try:
                            kcal_100g = float(kcal_100g) if kcal_100g else 0
                        except ValueError:
                            print(f"Valor no válido para 'kcal_100g': {kcal_100g}")
                            kcal_100g = 0
                        kcal_total += (amount * kcal_100g) / 100

                elif unit == 'unit/s':
                    # Buscar en food para obtener g_unit
                    food_query = food_ref.where('_id', '==', item_id).stream()
                    for food_doc in food_query:
                        food_data = food_doc.to_dict()
                        g_unit = food_data.get('g_unit', 0)
                        try:
                            g_unit = float(g_unit) if g_unit else 0
                        except ValueError:
                            print(f"Valor no válido para 'g_unit': {g_unit}")
                            g_unit = 0
                        kcal_total += amount * g_unit

                else:
                    # Buscar en utils_cons para obtener g_eq
                    utils_query = utils_cons_ref.where('nameEn', '==', unit).stream()
                    for utils_doc in utils_query:
                        utils_data = utils_doc.to_dict()
                        g_eq = utils_data.get('g_eq', 0)
                        try:
                            g_eq = float(g_eq) if g_eq else 0
                        except ValueError:
                            print(f"Valor no válido para 'g_eq': {g_eq}")
                            g_eq = 0
                        kcal_total += amount * g_eq

            # Actualizar la receta con el valor de kcal calculado
            recipes_ref.document(recipe_doc.id).update({'kcal': kcal_total})
            print(f"Receta con _id = {recipe_id} actualizada: kcal = {kcal_total}")

    except Exception as e:
        print(f"Error al calcular kcal: {e}")

# Llamar a la función
calculate_kcal_for_recipes()
```

```python
def calculate_kcal_adjusted_by_people():
    try:
        # Referencia a la colección 'recipes'
        recipes_ref = db.collection('recipes')

        # Obtener todas las recetas
        recipes_docs = recipes_ref.stream()

        for recipe_doc in recipes_docs:
            recipe_data = recipe_doc.to_dict()
            recipe_id = recipe_data.get('_id')
            kcal = recipe_data.get('kcal', 0)
            people = recipe_data.get('people', 1)

            # Validar que people sea un número y mayor a 1
            try:
                people = int(people) if people else 1
            except ValueError:
                print(f"Valor no válido para 'people' en receta con _id = {recipe_id}: {people}")
                people = 1

            if people > 1:
                kcal_adjusted = kcal / people
                # Actualizar el valor de kcal ajustado en la receta
                recipes_ref.document(recipe_doc.id).update({'kcal': kcal_adjusted})
                print(f"Receta con _id = {recipe_id} ajustada: kcal = {kcal_adjusted}, people = {people}")
            else:
                print(f"Receta con _id = {recipe_id} no necesita ajuste: kcal = {kcal}, people = {people}")

    except Exception as e:
        print(f"Error al ajustar kcal por people: {e}")

# Llamar a la función
calculate_kcal_adjusted_by_people()
```

```python
def calculate_time_mins_for_recipes():
    try:
        # Referencias a las colecciones
        recipes_ref = db.collection('recipes')
        recipes_actions_ref = db.collection('recipes_actions')

        # Obtener todas las recetas
        recipes_docs = recipes_ref.stream()

        for recipe_doc in recipes_docs:
            recipe_data = recipe_doc.to_dict()
            recipe_id = recipe_data.get('_id')
            time_mins = 0  # Inicializar el campo time_mins en 0

            print(f"\nCalculando tiempo para receta con _id = {recipe_id}")

            # Obtener las acciones relacionadas con la receta
            actions_query = recipes_actions_ref.where('id_recipe', '==', recipe_id).stream()

            for action_doc in actions_query:
                action_data = action_doc.to_dict()
                unit_time = action_data.get('unit_time')
                time_value = action_data.get('time')

                # Validar que unit_time y time_value sean correctos
                if unit_time and time_value:
                    # Normalizar el valor eliminando espacios y pasando a minúsculas
                    unit_time = unit_time.strip().lower()

                    try:
                        time_value = float(time_value)
                    except ValueError:
                        print(f"Valor no válido para 'time': {time_value} en receta {recipe_id}")
                        continue

                    # Sumar tiempo según la unidad corregida
                    if unit_time == "minute" or unit_time == "minutes":
                        time_mins += time_value
                    elif unit_time == "hour" or unit_time == "hours":
                        time_mins += time_value * 60
                    elif unit_time == "second" or unit_time == "seconds":
                        time_mins += time_value / 60
                    else:
                        print(f"Unidad de tiempo no reconocida: '{unit_time}' en receta {recipe_id}")
                else:
                    print(f"Acción sin 'unit_time' o 'time' en receta {recipe_id}, se omite.")

            # Actualizar el campo 'time_mins' en la receta
            recipes_ref.document(recipe_doc.id).update({'time_mins': time_mins})
            print(f"Receta con _id = {recipe_id} actualizada: time_mins = {time_mins} minutos")

    except Exception as e:
        print(f"Error al calcular 'time_mins': {e}")

# Llamar a la función
calculate_time_mins_for_recipes()
```

```python
def update_nutrients_for_all_recipes():
    try:
        # Referencias a las colecciones
        recipes_ref = db.collection('recipes')
        recipes_ing_ref = db.collection('recipes_ing')
        food_ref = db.collection('food')
        utils_cons_ref = db.collection('utils_cons')

        # Obtener todas las recetas
        recipes_docs = recipes_ref.stream()

        for recipe_doc in recipes_docs:
            recipe_data = recipe_doc.to_dict()
            recipe_id = recipe_data.get('_id')

            print(f"Recalculando nutrientes para receta con _id = {recipe_id}")

            # Obtener el número de personas, por defecto 1
            people = recipe_data.get('people', 1)
            try:
                people = int(people) if people else 1
            except ValueError:
                people = 1

            # Inicializar valores de nutrientes
            kcal_total = 0
            fats_total = 0
            protein_total = 0
            carbs_total = 0

            # Obtener los ingredientes de la receta
            ingredients_query = recipes_ing_ref.where('id_recipe', '==', recipe_id).stream()

            for ing_doc in ingredients_query:
                ing_data = ing_doc.to_dict()
                amount = ing_data.get('amount', 0)
                unit = ing_data.get('unit')
                item_id = ing_data.get('item_id')

                # Validar que el campo amount sea convertible a float
                try:
                    amount = float(amount) if amount else 0
                except ValueError:
                    print(f"Valor no válido para 'amount': {amount}")
                    continue

                # Buscar en food para obtener valores nutricionales
                food_query = food_ref.where('_id', '==', item_id).stream()
                for food_doc in food_query:
                    food_data = food_doc.to_dict()

                    kcal_100g = float(food_data.get('kcal_100g', 0) or 0)
                    fats_100g = float(food_data.get('fats', 0) or 0)
                    protein_100g = float(food_data.get('protein', 0) or 0)
                    carbs_100g = float(food_data.get('carbohydrates', 0) or 0)
                    g_unit = float(food_data.get('g_unit') or 0)  # Manejar si g_unit está vacío

                    if unit == 'g':
                        kcal_total += (amount * kcal_100g) / 100
                        fats_total += (amount * fats_100g) / 100
                        protein_total += (amount * protein_100g) / 100
                        carbs_total += (amount * carbs_100g) / 100
                    elif unit == 'unit/s':
                        kcal_total += amount * g_unit
                        fats_total += amount * g_unit
                        protein_total += amount * g_unit
                        carbs_total += amount * g_unit
                    else:
                        utils_query = utils_cons_ref.where('nameEn', '==', unit).stream()
                        for utils_doc in utils_query:
                            utils_data = utils_doc.to_dict()
                            g_eq = float(utils_data.get('g_eq', 0) or 0)
                            kcal_total += amount * g_eq
                            fats_total += amount * g_eq
                            protein_total += amount * g_eq
                            carbs_total += amount * g_eq

            # Dividir los valores de nutrientes por el número de personas
            kcal_total /= people
            fats_total /= people
            protein_total /= people
            carbs_total /= people

            # Actualizar la receta con los valores recalculados
            recipes_ref.document(recipe_doc.id).update({
                'kcal': kcal_total,
                'fats': fats_total,
                'protein': protein_total,
                'carbohydrates': carbs_total
            })

            print(f"Receta con _id = {recipe_id} actualizada: kcal = {kcal_total}, fats = {fats_total}, protein = {protein_total}, carbs = {carbs_total}")

    except Exception as e:
        print(f"Error al calcular nutrientes: {e}")

# Llamar a la función
update_nutrients_for_all_recipes()
```
