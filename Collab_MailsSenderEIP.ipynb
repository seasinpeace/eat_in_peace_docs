```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# Configuración del correo
gmail_user = 'seasinpeace@gmail.com'
gmail_password = 'czqb htkl vcbc kowq'

def enviar_correo(destinatario, asunto, mensaje, bcc=[]):
    msg = MIMEMultipart()
    msg['From'] = 'Eat In Peace <no-reply@eatinpeace.com>'
    msg['To'] = destinatario
    msg['Subject'] = asunto

    # ✅ Asegurarse de usar UTF-8 para los caracteres especiales
    msg.attach(MIMEText(mensaje, 'plain', 'utf-8'))

    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(gmail_user, gmail_password)
        server.sendmail(gmail_user, [destinatario] + bcc, msg.as_string())
        server.quit()
        print(f'Correo enviado a {destinatario}')
    except Exception as e:
        print(f'Error al enviar el correo: {e}')
```
